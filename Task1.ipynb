```python
import subprocess
import os

#Define the base directory where you want to store the repositories
base_dir = os.path.join(os.getcwd(), "repos")

#Check if the directory exists, if not create it
if not os.path.exists(base_dir):
    os.makedirs(base_dir)

#Define a list of repositories to download
repos = [
    "matplotlib/matplotlib",
    "scikit-learn/scikit-learn",
    "numpy/numpy",
    "pandas-dev/pandas",
    "django/django",
    "scipy/scipy",
    "pallets/flask",
    "psf/requests",
    "ansible/ansible",
    "getsentry/sentry",
    "scrapy/scrapy",
    "mailpile/Mailpile",
    "sshuttle/sshuttle",
    "saltstack/salt",
    "samuelclay/NewsBlur",
    "beetbox/beets"
]

for repo in repos:
    #Create the full repo URL
    repo_url = f"https://github.com/{repo}.git"

    #Create the directory path
    dir_path = os.path.join(base_dir, repo.split('/')[1])

    #Clone the repository
    subprocess.run(["git", "clone", repo_url, dir_path])
```
